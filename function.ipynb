```python
def cal(y):
    l,b,h=y
#     l=y[0]
#     b=y[1]
#     c=y[2]
    a=l*b
    v=a*h
    return [a,v] # return as list
l=int(input('enter l= '))
b=int(input('enter b= '))
h=int(input('enetr h= '))
x=(l,b,h) # for lots of data tuple is use for arguments passing because to incresase execution time
ar=cal(x)
area,volume=ar

print('the area is =',area)
print('the vlume is= ',volume)
```

```python
def cal(y):
    l,b,h=y
#     l=y[0]
#     b=y[1]
#     c=y[2]
    a=l*b
    v=a*h
    return {'area':a,'volume':v} # return dictionary
l=int(input('enter l= '))
b=int(input('enter b= '))
h=int(input('enetr h= '))
x=(l,b,h) # for lots of data tuple is use for arguments passing because to incresase execution time
ar=cal(x)
#area,volume=ar

print('the area is =',ar['area'])
print('the vlume is= ',ar['volume'])
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def err():
    print('value of b cnat be zero')
def cal():
    a=int(input('enter a= '))
    b=int(input('enter b= '))
    o=input('enter + - * /= ')
    if o=='+':
        print(add(a,b))
    elif o=='-':
        print(sub(a,b))
    elif o=='*':
        print(mul(a,b))
    elif o=='/':
        if b!=0:
            print(div(a,b))
        else:
            err()
    else:
        print('invalid operator')
cal()
```
