```python
#DictReader(csv ma vako data diction ma read garda use hunxa)
import csv
csvfile=open('books.csv')
reader=csv.DictReader(csvfile) # scv package use garera books.csv lai read gara
for row in reader:
    print(dict(row))
```

```python
import csv
csv_colums=['name','age','address']
dict=[
    {'name':'belmata','age':24,'address':'dang'},
    {'name':'asha','age':25,'address':'ktm'}
    
]

csvfile=open('data.csv','w')
writer=csv.DictWriter(csvfile,fieldnames=csv_colums)# fieldname is keyword
writer.writeheader()
for data in dict:
    writer.writerow(data)
```
